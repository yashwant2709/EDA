**Create data frame using dictinary**

```python
name=['Yashwant','Madhuri','Mokshita','tanisha','lavi']
age=[25,19,16,14,12,4]
pd.DataFrame(zip(name,age),columns=['Names', 'Ages'])
```

```python
dict1={'Names':['Yashwant','Madhuri','Mokshita','tanisha','lavi'],'Ages':[25,19,16,12,4]}
dict1
```

```python
pd.DataFrame(dict1)
# keys becomes columns
# Values becomes rows
# No need to provide column names seperately
```
