**Normalization**

- Read the data again

- step-1: take prevailing wage column : **visa_df['prevailing_wage']**

- step-2: Calculate the min value of prevailing wage:min=**visa_df['prevailing_wage'].min()**

- step-3: Calculate the max value of prevailing wage:max= **visa_df['prevailing_wage'].max()**

- step-4: Calculate the Nr: step1-step2 :Nr: **visa_df['prevailing_wage']-mean**

- step-4: Calculate the Nr: step1-step2 :Nr: **visa_df['prevailing_wage']-mean**

**package name: MinMaxScalar**

```python
x_max=visa_df['prevailing_wage'].max()
x_min=visa_df['prevailing_wage'].min()
Nr = visa_df['prevailing_wage']-x_min
visa_df['prevailing_wage_min_max']=Nr/(x_max-x_min)
visa_df[['prevailing_wage','prevailing_wage_min_max']]
```
