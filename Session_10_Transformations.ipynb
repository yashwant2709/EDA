**Data Transformations**

- Generally data has 3 types of distribution 

    - Normal distribution

    - Positive skewed

    - Negative skewed

- All the maths concepts developed by assumption as Data follows normal distribution

- Data skews happens because of many reasons , in that one reason is outliers

- One we perform the outliers analysis then we can apply transformation methods, if the data does not follows normality

- We have some transformation methods mentioned below:

   - Log Transformation

   - Exponential Transformation

   - Reciprocal Transformation

   - Square root Transformation

   - Power Transformation

$step-1$

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

$Step-2$

**Read the data**

```python
exp_data=np.random.exponential(size=10000)
exp_data
```

$Step-3$

**Plot the data**

```python
plt.hist(exp_data,bins=20,label='Exponential')    # bins means interval
plt.show()
```

```python
norm_data=np.random.normal(size=10000)
plt.hist(norm_data,bins=20,label='Exponential')
plt.show()
```

$Step-4$

**Log Transformation**

- Log transformation means performing logarithmic operations on original data

- It is an approach to convert into normal distribution 

- Log means natural logarithm base=e

```python
x=2
np.log(x)
```

```python
log_data=np.log(exp_data)
log_data
```

```python
log_data=np.log(exp_data)
plt.hist(log_data,bins=20,label='Logarithm')
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.subplot(1,2,1).hist(exp_data,bins=30,label="Exponential")
plt.legend()
plt.subplot(1,2,2).hist(log_data,bins=30,label="Logarithm")
plt.legend()
plt.show()
```

**Reciprocal Transformation**

- Reciprocal will fail when values has very less

- For example x=0 then 1/0 fail

```python
x=2
1/x
```

```python
reciprocal_data=np.reciprocal(exp_data)
plt.hist(reciprocal_data,bins=20,label='Reciprocal')
plt.legend()
plt.show()
```

**Square root transformation**

```python
square_root_data=np.sqrt(exp_data)
plt.hist(square_root_data,bins=20,label='Square root')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.subplot(1,3,1).hist(exp_data,bins=20,label="Exponential")
plt.legend()
plt.subplot(1,3,2).hist(log_data,bins=20,label="Logarithm")
plt.legend()
plt.subplot(1,3,3).hist(square_root_data,bins=20,label='Square root')
plt.legend()
plt.show()
```
